```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\MSI\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.rugplot(x='salary', data=df, height=0.25)
```

```python
#sns.set(style='darkgrid') #darkgrid, whitegrid, dark, white, ticks

sns.displot(data=df,x='salary', bins=20,color='red', 
            edgecolor='blue', ls='--', kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages= pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
sns.displot(data=sample_ages,x='age',rug=True,bins=30, kde= True)
```

```python
sns.kdeplot(data=sample_ages, x='age',clip=[0,100], bw_adjust=0.6,shade=True)
```
