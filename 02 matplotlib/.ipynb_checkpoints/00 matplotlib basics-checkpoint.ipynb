```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y
```

```python
fig = plt.figure()

#large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('power of 4');


#small axes
axes2 = fig.add_axes([1,1,0.25,0.25])
axes2.plot(x,y)

axes2.set_xlim(1,2)
axes2.set_ylim(0,10)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('zoomed in');


plt.show()
```

```python
type(fig)
```

```python
fig = plt.figure(figsize=(2,2), dpi = 200)  #dots per inch

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

fig.savefig('new_fig.png', bbox_inches = 'tight')
```

```python
#subplots functionality
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)

fig.suptitle("fig level", fontsize=16)

fig.subplots_adjust(wspace=1, hspace=0.5)

fig.savefig('new_subplts.png', bbox_inches = 'tight')
```

```python
#styling
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label = 'X VS X')

ax.plot(x,x**2, label = 'X VS X^2')

ax.legend(loc=(1.1,0.5))
```

```python
#styling part 2
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x,x,color='purple', label = 'x vs x', lw=2, ls = '--',marker='+', markersize = 10)
#we can also pass rgb hex code
#lw is line width
ax.plot(x,x+1, color='red', label = 'x vs x+1', lw=2, marker= 'o')

ax.legend()
```
